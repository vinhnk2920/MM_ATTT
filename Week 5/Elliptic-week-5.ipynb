# Import library

```python
import sympy as sp
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Check a,b có hợp lệ không
def find_ab(a, b):
    if 4 * a**3 + 27 * b**2 == 0:
        return False
    return a, b
```

```python
# Check số nguyên tố
def is_prime(n):
    """Check if a number is prime."""
    if n <= 1:
        return False
    elif n <= 3:
        return True
    elif n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
```

```python
def elliptic_chart(points):
    # Extract x and y values from the points
    x_values = [point[0] for point in points]
    y_values = [point[1] for point in points]
    
    # Plot the points
    plt.plot(x_values, y_values, marker='o', linestyle='-')

    # Add labels and title
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Elliptic curve equation')

    # Show the plot
    plt.grid(True)
    plt.show()
```

```python
# % is modulo in python
def elliptic(p, a, b):
    points = []
    ab_result = find_ab(a,b)
    if ab_result is False:
        return "a, b không hợp lệ"
    for x in range(1, p):
        y_square = (x**3 + a*x + b) % p
        for y in range(1, int((p-1)/2)+1):
            if (y**2) % p == y_square:
                points.append([x, y])
                points.append([x,p-y])
    print("Danh sách các điểm được tạo ra:", points)
    point_count = len(points)+1
    if is_prime(point_count):
        print("\nCó ",point_count, " được tạo ra, tổng số điểm là số nguyên tố. Dưới đây là đồ thị của đường cong Elliptic tạo ra:\n")
        elliptic_chart(points)
    else:
        return "Tổng số điểm không phải số nguyên tố"
```

```python
elliptic(23, 2, 9)
```
